```python
df['объём_двиг_л'] = df['объём_двиг_л'].str.replace(',', '.').str.replace(' ', '')

def clean_volume(value):
    if isinstance(value, str):
        removed = re.sub(r'[^\d.]', '', value)
        removed = re.sub(r'(?i)\bturbo\b', '', removed)
        removed = removed.replace('/', '.')
        try:
            return removed
        except ValueError:
            return np.nan
    return value

df['объём_двиг_л'] = df['объём_двиг_л'].apply(clean_volume)

df['объём_двиг_л'].unique()
```

```python
def fillna_mean(colum, group):

### Функция заполнение пропущенных значений в выбранном столбце 
### на основе среднего значения по конкретному столбцу ###

### colum - столбец в котором заполняем пропуски
### group - столбец по которому групперуем данные 

    grouped = df.groupby(group)
    
    df[colum] = grouped[colum].transform(lambda x: x.fillna(x.mean()))
```

```python
def transmission(type_transmission):
    
    if type_transmission in ['Powershift S6', 'PowershifТS6', 'Tiptronic', 'ТipТronic', 'Steptronic', 'S-Тronic', 
                               'S-Тroic', 'S-troic', 'S-tronic', 'SТepТronic', '7 DCT', '7DCТ', '7DCT',
                             'CVT(вариатор)', 'CVТ', 'CVT (вариатор)', 'CVT(вариATор)', 'CVТ(вариатор)', 'CVТ(вАриАтор)',
                             '7 DSG', '7DSG4МoТion', 'CVT', '7DSG' , '6 DSG', '6DSG', '6DSG',
                             'PDK', 'PowershiftS6']:
        return 'other'
    
    elif type_transmission in ['Автомат', 'АT', 'A/T', 'AT (автомат)', 'АТ', 'АКПП', 'АТ (автомат)', '6 АТ', '6АТ', '6AT', 
                               '6 AT', '6 АТ', '6АT', 'АКПП6', '6Aатомат', '6Aвтомат', '6-ступАКПП', '6A', '6А', '8 АТ', 
                               '8AT', 'АКПП8', '8АКПП', '8-ступ АКПП', '8', '8 AT 4Motion', '8АТ4МoТion', '5АТ', '5AT',
                               '4 АТ', '4AT', 'АКПП4', '4АТ', '9АТ', '9AT', '7AT', '7АТ', '7АT', '8АТ', '12AT']:
        return 'AT'
    
    elif type_transmission in ['5 МТ', '5МТ', 'МКПП5', '5-ступМТ', '5M', '5М', '5MТ', '5MT', 'МТ', 'MТ', 'Мех.', 'МКПП', 
                               'МТ (механика)', 'Механическая', 'N', '6МТ', '6MТ', '6 MT', '6 МТ', '6М/Т', '6M/T', '6M/Т',
                               'МКПП4', '4MT', '6MT', 'AMT']:
        return 'MT'
    
    elif type_transmission in ['6AT/MT', '0', ' -', '-']:
        return np.nan
        
    else: return type_transmission

 
df['коробка_передач'] = df['коробка_передач'].apply(transmission)
df['коробка_передач'].unique()
```

```python
def top_table(x_value, y_value):
    
    """Функция создает сводную таблицу, суммирует данные и сортирует их по убыванию"""
    
    top = df.pivot_table(index=x_value, values=y_value, aggfunc = sum).sort_values(by=y_value, ascending=False) \
    .reset_index().head(10)
    return top
```

```python
def top_barplots(x_value,y_value,title,x_label,y_label):
    
    """На основе таблицы строит столбчатую диаграмму"""
    
    top = df.pivot_table(index=x_value, values=y_value, aggfunc = sum).sort_values(by=y_value, ascending=False) \
    .reset_index().head(10)
    
    sns.set_style("ticks",{'axes.grid' : True})
    
    plt.figure(figsize = (14,6))
    
    ax = sns.barplot(x=x_value, y=y_value, data=top, palette='Set2')
    
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

    plt.show()
```

### Объем продаж 

```python
def scatterplot(x_value, y_value, value, title, x_label, y_label):
    
    pivot_df = df.pivot_table(index=x_value, columns=y_value, values=value, aggfunc='sum')

    # Удаление нулевых значений
    pivot_df = pivot_df.replace(0, pd.NA).dropna(how='all')

    # Преобразование данных для параметра size
    sizes = pivot_df.values.flatten()

    # Создание scatter plot
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=pivot_df.unstack().reset_index(), x=x_value, y=y_value, size=sizes, sizes=(10, 1000),
               alpha=0.6, legend=False)
    plt.xticks(rotation=90)

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)

    plt.show()
```

### Конкуренты

```python
def figure_barplot(x_value,y_value,hue_value, title, x_label, y_label, legend):
    
    '''Функция создает сводную таблицу, а затем строит столбчатую диаграмму с третим измерением'''
    
    data = mercur_auto[hue_value].unique()

    mercur_data = df[df[hue_value].isin(data)]

    mercur_data_pivot = mercur_data.groupby([hue_value,x_value])[y_value].sum()
    mercur_data_pivot=mercur_data_pivot.reset_index().sort_values(by=y_value,ascending=False)

    plt.figure(figsize = (14,7))
    ax = sns.barplot(x=x_value, y=y_value, data=mercur_data_pivot, hue=hue_value, estimator=sum,
                palette='Set2')

    ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
    
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.legend(title=legend)
    
    plt.show()
```
